```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
Data = json.loads(r.text)
print(len(Data))
#print(Data[0])

for (k, v) in Data[0].items():
   print("Key: " + k)
   print("Value: " + str(v))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in Data:
    print("æ¨™é¡Œ:", d['title'])
    print("è²¼æ–‡æ™‚é–“:", d['createdAt'])
    print("ç•™è¨€äººæ•¸:", d['commentCount'])
    print("æŒ‰è®šäººæ•¸:", d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
# çœ‹æ¿æ–‡ç« ï¼š https://www.dcard.tw/_api/forums/[çœ‹æ¿]/posts?popular=[false:æœ€æ–°;true:ç†±é–€] 

popular_Count = len(Data)
popular_commentCount = 0
popular_likeCount = 0
for d in Data:
    popular_commentCount += int(d['commentCount'])
    popular_likeCount += int(d['likeCount'])

print("ç†±é–€çš„æ–‡ç« ")
print("å¹³å‡ç•™è¨€äººæ•¸:", (0 if (popular_Count<=0) else round((popular_commentCount/popular_Count),2)))
print("å¹³å‡æŒ‰è®šäººæ•¸:", ((0, round((popular_likeCount/popular_Count),2))[(popular_Count>0)]))
```
